```python
import requests
import json
import pandas as pd
```

```python
SeriesID = ["LNS14000012"]
SaveFileAs = "Unemployment Data_16-19.csv"
```

```python
#Get your jsonified api data here

#Created 3 different time frames because api.bls.gov only allows 20 years at a time.
startyear = [2000,1980,1960]
endyear = [2019,1999,1979]

#Empty lists for Years, Months, and Unemployment Rate
Years = []
Months = []
Unemployment_Rate = []

# Iterate over the 3 times (1960-1979, 1980-1999, 2000-2019)
for year in range(len(startyear)):
    # Base api link
    base_URL = "https://api.bls.gov/publicAPI/v2/timeseries/data/"
    
    # Elements for the api link
    headers = {'Content-type': 'application/json'}
    data = json.dumps({"seriesid": SeriesID,"startyear":startyear[year], "endyear":endyear[year],"registrationkey":"a1996666110d420cbf38f706289109f6"})
    
    # Request for api data
    Complete_Data = requests.post(f"{base_URL}",data=data, headers=headers)
    
    # jsonify the data
    Complete_Data = Complete_Data.json()
    
    # for loop to iterate through the api data and grab all the important information
    for i in Complete_Data["Results"]["series"][0]["data"]:
        Years.append(i["year"])
        Months.append(int(i["period"].replace('M', '')))
        Unemployment_Rate.append(i["value"])
        
Complete_Data
```

```python
Collected_Data = pd.DataFrame({"Years":Years,
                              "Months":Months,
                              "Unemployment Rate":Unemployment_Rate})

Collected_Data
```

```python
Collected_Data.to_csv(SaveFileAs,index=False)
```
